# Clean and Analyze Employee Exit Surveys
This is the sixth guided project of the Data Scientist Python Path in Dataquest and we are going to analyze two surveys:from the Department of Education, Training and Employment (DETE) and from the Technical and Further Education (TAFE). The DETE dataset can be found [here](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey) and the TAFE dataset can be found [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey).  
In this project we are going to answer two main questions:
* Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?
* Are younger employees resigning due to some kind of dissatisfaction? What about older employees?  

The objective is to combine both datasets to answer the data. The data also has no dictionary, so we stay with our definition of the headers of the columns.

    - `Contributing Factors. Dissatisfaction`

To do so, we are going to create the `dissatisfied` column, change the values in the columns to boolean type and check if any of the values in the correspondent columns return `True`, if so, we will mark the `dissatisfied` column as `True`.   

```python
# let's check the values of tafe_resignations
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
# let's check the values of tafe_resignations
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
# creating a function to convert the values to True or False
def update_vals(val):
    if pd.isnull(val):
        return (np.nan)
    elif val == '-':
        return (False)
    else:
        return True
```

```python
# using our function in the columns of tafe_resignations
tafe_resignations[['Contributing Factors. Job Dissatisfaction', 
                   'Contributing Factors. Dissatisfaction']] = tafe_resignations[['Contributing Factors. Job Dissatisfaction',
                                                                                  'Contributing Factors. Dissatisfaction']].applymap(update_vals)

tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(dropna=False)
```

```python
# creating the 'dissatisfied' column in tafe_resignations dataframe
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Job Dissatisfaction', 
                                                       'Contributing Factors. Dissatisfaction']].any(axis=1, skipna=False)

tafe_resignations['dissatisfied'].head()
```

```python
# creating the 'dissatisfied' column in dete_resignations dataframe
dete_resignations_nonull['dissatisfied'] = dete_resignations_nonull[['job_dissatisfaction', 
                                                                     'dissatisfaction_with_the_department',
                                                                     'physical_work_environment',
                                                                     'lack_of_recognition', 
                                                                     'lack_of_job_security',
                                                                     'work_location', 
                                                                     'employment_conditions',
                                                                     'work_life_balance', 
                                                                     'workload']].any(axis=1,
                                                                                      skipna=False)

dete_resignations_nonull['dissatisfied'].head()
```

```python
# creating copies of our dataframes
dete_resignations_up = dete_resignations_nonull.copy()
tafe_resignations_up = tafe_resignations.copy()

# checking if it worked
print('dete_resignations_up shape: {}'.format(dete_resignations_up.shape))
print('tafe_resignations_up shape: {}'.format(tafe_resignations_up.shape))
```

## 6 Merging our datasets
To do so, we need to add a column in both dataframes that comunicate which institute each survey is about, this column will be named `institute` in both dataframes, in dete the entry will be `DETE` and in tafe will be `TAFE`.

```python
# creating the columns and populating them
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'

print('dete institute: \n{}\n\ntafe institute: \n{}'.format(dete_resignations_up['institute'].value_counts(),
                                                      tafe_resignations_up['institute'].value_counts()))
```

Now that we created the columns, we are going to merge the dataframes and drop any columns with less than 500 non null values. We've chosen 500 based on the amount of rows each dataframe has.

```python
# merging the dataframes using pd.concat
combined = pd.concat([dete_resignations_up, tafe_resignations_up], axis=0)

# checking fast the null values per column
for column in combined.columns:
    print('{}:{}\n'.format(column, combined[column].isnull().sum()))
```

```python
# dropping the columns with less than 500 non null values
combined_updated = combined.dropna(axis=1, thresh=500).copy()

# checking fast the null values per column
for column in combined_updated.columns:
    print('{}:{}\n'.format(column, combined_updated[column].isnull().sum()))
```

Above we can see that there are several value types in this column, numbers and strings, with years and year ranges. So we are going to work on this column before our analysis.
## 7 Categorazing the `institute_service` column entries
To do so, we are going to categorize our `institute_service` column data creating categories based on a range of service years. This approach in based in [this article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage) that arguments the effectiveness of understanding employee's need based on career stage instead of age.  
We will follow it, but we will change the names of the categories, the years at a company per stage remains the same:  
- New: Less then 3 years  
- Experienced: 3 to 6 years
- Established: 7 to 10 years
- Veteran: 11 or more years  

First we need to extract the years of the columns, then create a function that categorizes the entries and at the end apply them in the entire column, creating another column to store the results.

```python
# converting the column to string type
combined_updated['institute_service'] = combined_updated['institute_service'].astype(str)

# creating the regular expression
year_pattern = r"(\d+)"

# extracting the string to another column
combined_updated['institute_service'] = combined_updated['institute_service'].str.extract(year_pattern, expand=True)

# turning the data into numbers again
combined_updated['institute_service'] = combined_updated['institute_service'].astype(float)

# cheking it out
combined_updated['institute_service'].value_counts(dropna=False).sort_index(ascending=True).head(15)
```

Everything seems to work fine, so let's make our special function.

```python
# defining our function
def career_stage(year):
    if pd.isnull(year):
        pass
    elif year <= 3:
        return "New"
    elif year <= 6:
        return "Experienced"
    elif year <= 10:
        return "Established"
    else:
        return "Veteran"
    
# testing it out
career_stage(12)
```

```python
# applying our function to the whole column
service_cat = combined_updated['institute_service'].apply(career_stage)

# checking it out
service_cat.value_counts(dropna=False)
```

Now that we finished it, we will come back to our `dissatisfied` column, fill the missing values and start our analysis.
## 8 Filling the missing values
To begin we are going to check the amount of missing values.

```python
# looking at the amount of true, false and null values
combined_updated['dissatisfied'].value_counts(dropna=False)
```

To fill our missing values we are going to use `False` since it is the most common value of the column.

```python
# filling it all out
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
combined_updated['dissatisfied'].value_counts(dropna=False)
```

After filling the missing values, we can start exploring our data.
## 9 Analyzing the data
Now that we finished part of the cleaning, we can start to make some analysis, to do so we are going to calculate percentages and plot some graphs to see the information.

```python
# using the pivot table to calculate percentage of each
pivot_combined = combined_updated.pivot_table(values='dissatisfied', index=service_cat)
pivot_combined
```

```python
# plotting the results
pivot_combined.plot(kind='bar', legend=False)
```
